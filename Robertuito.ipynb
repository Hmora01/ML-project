```python
import pandas as pd

# Cargar los archivos .xlsx para inspeccionar su contenido
experiencia_xlsx_path = '/content/Experiencia_2018_verano_2022.xlsx'
etiquetado_xlsx_path = '/content/Experiencia_etiquetado.xlsx'

# Leer los archivos Excel
try:
    experiencia_xlsx_df = pd.read_excel(experiencia_xlsx_path)
    etiquetado_xlsx_df = pd.read_excel(etiquetado_xlsx_path)
    # Leer exitosamente, mostrar las primeras filas
    experiencia_xlsx_head = experiencia_xlsx_df.head()
    etiquetado_xlsx_head = etiquetado_xlsx_df.head()
except Exception as e:
    # En caso de otro error, capturarlo
    experiencia_xlsx_head, etiquetado_xlsx_head = str(e), str(e)

(experiencia_xlsx_head, etiquetado_xlsx_head)
```

```python
import string

# Extraer la columna de "Comentarios y Sugerencias" de ambos DataFrames
comentarios_df = experiencia_xlsx_df['Comentarios y Sugerencias']
etiquetados_df = etiquetado_xlsx_df['Comentarios y Sugerencias']

# Ajustar la función de limpieza para manejar valores no textuales (como NaN)
def limpiar_comentario(texto):
    # Verificar si el valor es una cadena de texto
    if pd.isna(texto):
        return ""
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar signos de puntuación
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    # Eliminar números
    texto = ''.join([i for i in texto if not i.isdigit()])
    # Eliminar espacios adicionales
    texto = ' '.join(texto.split())
    return texto

# Aplicar la limpieza a cada comentario
comentarios_clean = comentarios_df.apply(limpiar_comentario)
etiquetados_clean = etiquetados_df.apply(limpiar_comentario)

# Mostrar los primeros 5 comentarios limpios para verificar
comentarios_clean.head(), etiquetados_clean.head()
```

```python
# Asegúrate de tener el archivo de palabras vacías accesible y actualiza la ruta de acceso si es necesario
lista_path = '/content/Lista.txt'

# Cargar la lista de palabras vacías
with open(lista_path, 'r', encoding='utf-8') as file:
    lista_palabras = file.readlines()

# Convertir la lista de palabras vacías a un conjunto para eliminar duplicados y búsquedas más rápidas
stop_words_sp = set(word.strip() for line in lista_palabras for word in line.split(',') if word.strip())

# Ahora, aquí sigue la definición de la función eliminar_palabras_vacias y su aplicación a los comentarios
```

```python
# Convertir la lista de palabras vacías a una lista de Python (limpiando y eliminando los caracteres innecesarios)
stop_words_sp = [word.strip() for line in lista_palabras for word in line.split(',') if word.strip() != '']
stop_words_sp = set(stop_words_sp)  # Convertir a un conjunto para una búsqueda más rápida

def eliminar_palabras_vacias(texto, stop_words):
    # Separar el texto en palabras
    palabras = texto.split()
    # Filtrar las palabras que no están en la lista de palabras vacías
    palabras_filtradas = [palabra for palabra in palabras if palabra not in stop_words]
    # Unir las palabras filtradas de nuevo en un texto
    texto_filtrado = ' '.join(palabras_filtradas)
    return texto_filtrado

# Limpiar la lista de palabras vacías quitando las comillas innecesarias
stop_words_sp = set(word.replace("'", "").strip() for word in stop_words_sp)

# Aplicar nuevamente la función de eliminación de palabras vacías a los comentarios
comentarios_sin_stopwords = comentarios_clean.apply(lambda x: eliminar_palabras_vacias(x, stop_words_sp))
etiquetados_sin_stopwords = etiquetados_clean.apply(lambda x: eliminar_palabras_vacias(x, stop_words_sp))

ejemplo_comentario = comentarios_clean.iloc[2]
# Mostrar las primeras 5 palabras vacías limpias y el ejemplo de comentario después de la limpieza
ejemplo_comentario_limpio = eliminar_palabras_vacias(ejemplo_comentario, stop_words_sp)

list(stop_words_sp)[:10], ejemplo_comentario_limpio
```

```python
import pandas as pd
import re
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from pysentimiento import create_analyzer

# Función para extraer el sentimiento y la mayor probabilidad de un string que representa un objeto AnalyzerOutput
def extract_sentiment_info_from_string(output_str):
    if pd.isna(output_str):
        return None, None
    sentiment_match = re.search(r"output=([A-Z]+)", output_str)
    probas_match = re.search(r"probas={([^}]+)}", output_str)
    if sentiment_match and probas_match:
        sentiment = sentiment_match.group(1)
        probas_str = probas_match.group(1)
        probas = dict((k.strip(), float(v.strip())) for k, v in
                      (pair.split(':') for pair in probas_str.split(',')))
        max_proba = max(probas.values())
        return sentiment, max_proba
    else:
        return None, None

# Cargar el archivo de comentarios con sentimientos
df_sentimientos = pd.read_csv("comentarios_con_sentimiento.csv")

# Aplicar la función para extraer la información de sentimiento y probabilidad
sentiment_info = df_sentimientos['Sentiment'].apply(lambda x: extract_sentiment_info_from_string(x))

# Crear nuevas columnas para el sentimiento y la probabilidad máxima
df_sentimientos['Sentiment_Clean'] = sentiment_info.apply(lambda x: x[0] if x else None)
df_sentimientos['Proba'] = sentiment_info.apply(lambda x: x[1] if x else None)

# Cargar el archivo Experiencia_2018_verano_2022.xlsx en un DataFrame
df_experiencia = pd.read_excel("Experiencia_2018_verano_2022.xlsx")

# Asegurarte de que los comentarios están en el mismo orden
df_sentimientos['ORGANIZACION'] = df_experiencia['ORGANIZACION']

# Agrupar los comentarios por empresa y contar las frecuencias de cada sentimiento
resultados_por_organizacion = df_sentimientos.groupby('ORGANIZACION')['Sentiment_Clean'].value_counts().unstack().fillna(0)

# Identificar las 5 organizaciones con más comentarios positivos, negativos y neutrales
top_positivas = resultados_por_organizacion['POS'].sort_values(ascending=False).head(5)
top_negativas = resultados_por_organizacion['NEG'].sort_values(ascending=False).head(5)
top_neutrales = resultados_por_organizacion['NEU'].sort_values(ascending=False).head(5)

# Imprimir los resultados
print("Organizaciones con más comentarios positivos:\n", top_positivas)
print("Organizaciones con más comentarios negativos:\n", top_negativas)
print("Organizaciones con más comentarios neutrales:\n", top_neutrales)
```
